```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Read the CSV file (skip first 4 extra lines – World Bank format)
df = pd.read_csv("/content/API_SP.POP.TOTL_DS2_en_csv_v2_2461.csv", skiprows=4)

# See first few rows
df.head()
```

```python
year = '2020'  # you can change to '2019', '2021', etc.

# Keep only Country Name and 2020 column, and remove missing values
data_2020 = df[['Country Name', year]].dropna()

data_2020.head()
```

```python
# Sort by population and take top 10 countries
top10 = data_2020.sort_values(by=year, ascending=False).head(10)

# Bar chart
plt.figure(figsize=(10, 6))
plt.bar(top10['Country Name'], top10[year])

plt.title('Top 10 Countries by Population in 2020')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45, ha='right')  # rotate names so they fit
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data_2020[year], bins=20)

plt.title('Distribution of Country Populations in 2020')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```
